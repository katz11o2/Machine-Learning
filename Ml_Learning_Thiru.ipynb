```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.title('combining all prerequists')
plt.show()
```

```python
import matplotlib.pyplot as plt
dar=plt.figure()
ax=dar.add_axes([0,0,1,1])
language=('tamil','kannada','hindi')
people=(100,20,30)
plt.bar(language,people)
```

```python
import matplotlib.pyplot as plt
dar=plt.figure()
ax=dar.add_axes([0,0,1,1])
language=('tamil','kannada','hindi')
people=(100,20,30)
plt.pie(people,labels=language,autopct='%1.1f%%')
```

```python
#3d plot
import matplotlib.pyplot as plt
import numpy as np
dx=plt.axes(projection='3d')
x=np.linspace(-10,10,200)
y=np.sin(x)
z=np.cos(x)
plt.scatter(x,y,z,c=z,cmap='Blues')
```
